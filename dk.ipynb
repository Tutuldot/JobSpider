```python
# Define the header row
header = [
    "LiberPage",
    "Grantor",
    "Grantee",
    "LegalNotes",
    "DocumentID",
    "documentUrl",
    "DocumentNumber",
    "Recording Date",
    "Consideration"
]

# Define the output file name
output_file_name = "data.csv"

# Check if the file exists
if not os.path.exists(output_file_name):
    # Create the file and write the header
    with open(output_file_name, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile, delimiter=';')
        writer.writerow(header)

def write_data(dictEntry):
    with open(output_file_name, 'a', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=header, delimiter=';')
        writer.writerow(dictEntry)


def generateDictionary(liDoc):
    liberPage = []
    Grantor = []
    Grantee = []
    LegalNotes = []
    documentId = None
    documentUrl = None
    documentNumber = None
    recordedDate = None

    mainLi = liDoc
    documentId = mainLi.get('data-documentid')
    documentUrl = mainLi.get('href')
    lilist = mainLi.select('li')

    pattern = r"\b\d{10}\b"

    match = re.search(pattern, str(mainLi))

    if match:
        documentNumber = match.group()

    pattern = r"\d{2}/\d{2}/\d{4} \d{2}:\d{2} (AM|PM)"

    match = re.search(pattern, str(mainLi))

    if match:
        recordedDate = match.group()


    isHeader = False
    headerType = ""
    for li in lilist:

        content = str(li.contents[0])
        content = content.replace("<b>","").replace("</b>","")
        potentialHeader = ["Liber/Page","Grantor","Grantee","Legal/Notes"]
        
        for ph in potentialHeader:
            if ph in content:  
                headerType = ph
                isHeader = True
                # print(f"header: {ph}")
                break
            else: 
                isHeader = False

        # header is all set

        if not isHeader:
            # print(f"Content: {content} stored as : {headerType}")
            if headerType == "Liber/Page" :
                liberPage.append(content)
            elif headerType == "Grantor":
                Grantor.append(content)
            elif headerType == "Grantee":
                Grantee.append(content)
            elif headerType == "Legal/Notes":
                LegalNotes.append(content)
        

    finalLiberPage = ", ".join(liberPage)
    finalGrantor = ", ".join(Grantor)
    finalGrantee = ", ".join(Grantee)
    finalLegalNotes = ", ".join(LegalNotes)

    finalEntry = {
        "LiberPage" : finalLiberPage
        ,"Grantor" : finalGrantor
        ,"Grantee" : finalGrantee
        , "LegalNotes" : finalLegalNotes
        , "DocumentID" : documentId
        , "documentUrl" : "https://waynecountymi-web.tylerhost.net/web/document/" + documentId
        , "DocumentNumber" : documentNumber
        , "Recording Date" : recordedDate
        , "Consideration" : "NA"

    }

    

    return finalEntry
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument( '--disable-blink-features=AutomationControlled' )
driver = webdriver.Chrome(options=chrome_options)
```

```python
driver.get("https://waynecountymi-web.tylerhost.net/web/search/DOCSEARCH1906S1")


```

```python
start_page = 22
last_page = 102

for i in range(start_page, last_page):
    # do something

    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'lxml')
    xc = soup.select('li[data-documentid]')
    for x in xc:
        dataDict = generateDictionary(x)
        write_data(dataDict)

    for jk in range(1,100):
        print("press down")
        driver.find_element(By.TAG_NAME, "body").send_keys(Keys.ARROW_DOWN)

    next_anchors = driver.find_elements(By.XPATH, "//a[@class='ui-link ui-btn ui-btn-b ui-btn-inline ui-shadow ui-corner-all' and @data-iconpos='right' and @data-inline='true' and text()='Next']")
    

   
    for b in range(1,10):
        error_occurred = False
        try:
            for anchor in next_anchors:
                anchor.click()
        except Exception as e:
            # print("err")
            error_occurred = True

        if not error_occurred:
            break

        time.sleep(1)
    
    print(f"done: {i}")
    time.sleep(2)
    
```
